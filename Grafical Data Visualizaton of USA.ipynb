```python
import chart_studio.plotly as py
import plotly.graph_objs as go
import pandas as pd
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
init_notebook_mode(connected = True)
```

```python
data = dict(type = "choropleth",
          locations = ["AZ","CA","NY"],
          locationmode = "USA-states",
          colorscale = "Viridis",
           text = ["Arizona","California", "New York"],
          z= [10,20,40],
          colorbar = {"title":"colorbar title goes here"})              # here in text the thing we will write we will get in the color plot for example if we write text 1 then it will show text 1 and if we write californa then it will show calefornia
```

```python
layout = dict(geo= {"scope":"usa"})
```

```python
choromap = go.Figure(data = [data],layout = layout)
```

```python
iplot(choromap)
```
